```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
data = {'Algorithm':['Perceptron','Multilayer Perceptron','ANN','Naïve Bayes','Random Forest','Logistic Regression','k-NN','SVM', 'Decision Tree','Hybrid Model'],
       'Our_Model':['67.0','83.9','NA','77.2','83.7','78.5','88.3','87.0','85.7','90.6'],
       'Surveyed_Paper':['NA','NA','88.57','78.28','77.14','78.85','79.42','77.71','73.14','NA']}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot()
ax.plot(df.Algorithm,df.Surveyed_Paper, color = 'red')
ax.plot(df.Algorithm,df.Our_Model,  color = 'blue')
ax.set_xlabel('Algorithms Used')
ax.set_ylabel('Accuracy in %')
ax.set_title('Comparision With Existing System')
ax.legend(['Existing System','Our Model'],loc='upper left')
ax.grid()
fig.savefig('Graph.png', bbox_inches='tight')
```
