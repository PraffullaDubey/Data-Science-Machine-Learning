```python
data['Province/State'].fillna('NA',inplace=True)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
leo=LabelEncoder()
```

```python
genre_label=leo.fit_transform(data['Province/State'])
genre_label
```

```python
en=leo.classes_
en
genre_map={index:label for index ,label in enumerate(en)}
genre_map
```

```python
#Adding the new values into dataframe/dataset
data['State/Province']=genre_label
data.drop('Province/State',inplace=True,axis=1)
data.head()
```

```python
genre_label=leo.fit_transform(data['Country/Region'])
genre_label
```

```python
en=leo.classes_
en
genre_map={index:label for index ,label in enumerate(en)}
genre_map
```

```python
#Adding the new values into dataframe/dataset
data['Region/Country']=genre_label
data.drop('Country/Region',inplace=True,axis=1)
data.head()
```
